Replacing Elements in NumPy Array



From the given NumPy array, replace all elements that are less than 'm' with 'm', and replace all elements that are greater than 'n' with 'n' where 'm' and 'n' are integer values provided as input.



Input format:

A NumPy array (as a list) on line 1

Integer m on line 2

Integer n on line 3



Output format:

A NumPy array where:

All elements < m are replaced with m

All elements > n are replaced with n



Test Case 1:

Sample input:

[2, 8, 15, 5, 20, 11, 3]

6

12



Sample output:

[6 8 12 6 12 11 6]



Test Case 2:

Sample input:

[10, 12, 14, 16]

12

15



Sample output:

[12 12 14 15]



```python
import numpy as np
import ast

# Read and convert input list into NumPy array
arr = np.array(ast.literal_eval(input().strip()))

# Read m and n
m = int(input().strip())
n = int(input().strip())

# Write your logic below
#arr = np.clip(arr, m, n)
arr[arr < m] = m
arr[arr > n] = n
# Print the resulting NumPy array
print(arr)
```
