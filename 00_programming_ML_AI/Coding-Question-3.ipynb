Extracting Elements at Even Indices



From the given NumPy array, extract all elements that are located at even indices (0, 2, 4, …).



Input Format

A NumPy array (as a list) on line 1



Output Format

A NumPy array containing elements from all even index positions of the original array



Test Case 1

Sample Input

[4, 7, 9, 12, 15, 3, 8]



Sample Output

[4 9 15 8]



Test Case 2

Sample Input

[10]



Sample Output

[10]

```python
import numpy as np
import ast

# Read and convert input list into NumPy array
arr = np.array(ast.literal_eval(input().strip()))

# Write your logic below
result = arr[::2]
# Print the resulting NumPy array
print(result)
```
