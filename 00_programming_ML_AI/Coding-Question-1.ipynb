Instructions

Extracting Elements from Array



From a given array, extract all the elements which are greater than 'm' and less than 'n' where 'm' and 'n' are integer values provided as input.



Input format

A list of integers on line one

Integer 'm' on line two

Integer 'n' on line three



Output format

1-D array containing integers greater than 'm' and smaller than 'n'



Test Case 1

Sample input

[1, 5, 9, 12, 15, 7, 12, 9]

6

12



Sample output

[9 7 9]



Test Case 2

Sample input

[1, 5, 9, 12, 15, 7, 12, 9]

12

6



Sample output

[]

```python
import ast
import numpy as np
list_of_integers = ast.literal_eval(input("Enter the list of integers: "))
print(list_of_integers)
int_arr = np.array(list_of_integers)
int_arr.dtype
m = int(input("Enter the lower bound:"))
n = int(input("Enter the upper bound:"))
filter_mask = ((int_arr > m) & (int_arr < n))
int_arr[filter_mask]
print(int_arr[filter_mask])
```
